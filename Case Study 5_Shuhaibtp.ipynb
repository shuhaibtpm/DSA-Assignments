### Hypothesis Test

H0: All the zones generated same amount of sales

H1: All the zones did not generate same amount of sales

```python
f_statistic, p_value = stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'], data['Zone - D'])
f_statistic,p_value
```

```python
if p_value<0.05:
    print('Reject the null hypothesis')
else:
    print("Accept the null hypothesis")
```

### here null hypothesis rejected hence All the zones  generate different amount of sales
