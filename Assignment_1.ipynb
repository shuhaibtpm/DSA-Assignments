# Calculating mean values of columns

```python
# here calculating the mean values of SL, SW, PL and PW
print('Mean value of sepal length(SL) is - ', iris['SL'].mean())
print('Mean value of sepal width(SW) is  - ', iris['SW'].mean())
print('Mean value of petal length(PL) is - ', iris['PL'].mean())
print('Mean value of petal width(PW) is  - ', iris['PW'].mean())
```

# Identifying null values

Two common methods to indicate where values in a DataFrame are missing are `isna` and `isnull`. They are exactly the same methods, but with different names.

```python
iris['SL'].isna()
```

```python
iris['SL'].isnull()
```

Here `False` indicating the value is not null value.
We can also use `info()` method to find the presence of null values in the dataset

```python
iris.info()
```

From this summary table we can see that out of 150 entries there is no null values

# Data Visualization

# 1. Histogram

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.hist(iris['SL'],rwidth=0.9, density=True)
plt.xlabel('Sepal Length', fontsize=16)
plt.ylabel('Frequency',fontsize=14)

plt.subplot(2,2,2)
plt.hist(iris['SW'],rwidth=0.9, density=True)
plt.xlabel('Sepal Width', fontsize=16)
plt.ylabel('Frequency',fontsize=14)

plt.subplot(2,2,3)
plt.hist(iris['PL'],rwidth=0.9, density=True)
plt.xlabel('Petal Length', fontsize=16)
plt.ylabel('Frequency',fontsize=14)

plt.subplot(2,2,4)
plt.hist(iris['PW'],rwidth=0.9, density=True)
plt.xlabel('Petal Width', fontsize=16)
plt.ylabel('Frequency',fontsize=14)

#Add a title for the entire figure
plt.suptitle('Histograms for the four features', fontsize=16)

plt.tight_layout()
```

# 2. Heat Map

```python
data = iris.corr()
plt.subplots(figsize=(8,6))
sns.heatmap(data, annot = True, square = True)
plt.title('Histogram', fontsize=16)
```

From above heatmap we can see that petal measurements have highly positive correlation, while the sepal one are uncorrelated. also the petal features also have relatively high correlation with sepal length, but not with sepal width

# 3. Strip Plot

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.stripplot(x=iris['SL'])
plt.xlabel('Sepal Length', fontsize=16)

plt.subplot(2,2,2)
sns.stripplot(x=iris['SW'])
plt.xlabel('Sepal Width', fontsize=16)

plt.subplot(2,2,3)
sns.stripplot(x=iris['PL'])
plt.xlabel('Petal Length', fontsize=16)

plt.subplot(2,2,4)
sns.stripplot(x=iris['PW'])
plt.xlabel('Petal Width', fontsize=16)

#Add a title for the entire figure
plt.suptitle('Strip Plot for the four features', fontsize=16)

plt.tight_layout()
```
