# Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load csv File

```python
df=pd.read_csv('cars_data.csv')
df.head()
```

# Check for the null values present in the dataset

```python
df.isnull().sum()
```

No null values present in the dataset

# Plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x=df['Buyer Gender'],
                   order=df['Buyer Gender'].value_counts(ascending=False).index);
        
abs_values = df['Buyer Gender'].value_counts(ascending=False)
rel_values = df['Buyer Gender'].value_counts(ascending=False, normalize=True).values * 100
lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]

ax.bar_label(container=ax.containers[0], labels=lbls,fontsize=16)
plt.title('Male vs Female Participation', fontsize=22,fontweight="bold")
plt.xlabel('Gender', fontsize=16,fontweight="bold")
plt.ylabel('Participation Count', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16)
plt.show()
```

From above graph it is clearly showing that female participation in the sale is slightly higher than the male.
